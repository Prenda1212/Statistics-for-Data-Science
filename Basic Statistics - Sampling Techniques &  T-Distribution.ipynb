Similar to a z-score table used with a normal distribution, a t-distribution uses a t-table. Knowing the degrees of freedom and the desired cumulative probability (e.g. P(T >= t) ) we can find the value of t. Here is an example of a lookup table for a t-distribution


```python
from scipy.stats import t

#create ta list of value
x = np.linspace(-4,4,100)

#create t-distribution.
t_1=t(10)

#plot probability density functions
plt.plot(x,t_1.pdf(x))
```

```python
#t-distribution
rv3=t(3)
rv10=t(10)
rv15=t(15)

#normal distribution
mean = 0
std = 1
X = np.arange(-4,4,0.01)
#Using (range of input, mean, standard diviation)
Y = stats.norm.pdf(X,mean,std)

#Plotting
plt.plot(x,rv3.pdf(x), label='t, df=3')
plt.plot(x,rv10.pdf(x), label='t, df=10')
plt.plot(x,rv15.pdf(x), label='t, df=15')
plt.plot(X,Y, label='normal')

plt.legend()
```
